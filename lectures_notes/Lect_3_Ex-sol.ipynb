# Exercise 3.4 solution

Run the following simulation $200$ times:
- $x_i\sim \mathcal{N}(\mu_i,1)$ for $i=1,2,...,500$.
- $\mu_i =3i/500$
- $i_{max}=$ index of largest $x_i$
- $d=x_{i_{max}}-\mu_{i_{max}}$

(b) Plot the histogram of the $200$ $d$ values. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
n_simulations = 200
n_values = 500

# Define the mean values for each x_i
mu_values = [3*i/500 for i in range(1, n_values + 1)]

# Initialize a list to store the d values
d_values = []

# Run the simulation
for _ in range(n_simulations):
    x_values = np.random.normal(mu_values, 1)
    i_max = np.argmax(x_values)
    d = x_values[i_max] - mu_values[i_max]
    d_values.append(d)

    

# Plot the histogram
plt.hist(d_values, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('d values')
plt.ylabel('Frequency')
plt.title('Histogram of d values from 200 simulations')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
